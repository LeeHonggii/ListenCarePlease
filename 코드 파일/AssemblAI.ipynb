```python
import assemblyai as aai
import os
# from google.colab import drive # Colab 사용자용: 주석 해제하여 Google Drive 마운트

# ⚠️ 1. API 키 설정 (보안에 유의) ⚠️
# 실제 키를 사용하셨습니다.
MY_API_KEY = "145408260d3346a09b5ba862da08d576"
aai.settings.api_key = MY_API_KEY

# Colab 사용자: 이 부분을 실행하여 Google Drive를 마운트해야 파일 접근이 가능합니다.
# drive.mount('/content/drive')

def transcribe_with_speaker_diarization_korean(audio_file_path):
    """
    AssemblyAI를 사용하여 오디오 파일에 대한 한국어 전사, 화자 분리 및 화자 수 카운트를 수행합니다.
    """

    # 2. 한국어 및 화자 분리 설정
    config = aai.TranscriptionConfig(
        speaker_labels=True,      # 화자 분리 활성화
        language_code="ko"        # 한국어 코드 지정
    )

    transcriber = aai.Transcriber()
    print(f"'{audio_file_path}' 파일 전사를 시작합니다. (언어: 한국어, 화자 분리 활성화)")

    # 3. 파일 전사 및 화자 분리 수행
    transcript = transcriber.transcribe(
        audio_file_path,
        config=config
    )

    # 4. 결과 확인 및 화자 수 카운트
    if transcript.status == aai.TranscriptStatus.error:
        print(f"\n🚨 전사 오류 발생: {transcript.error}")
        return

    # 모든 발화(utterance)에서 화자 레이블만 추출하고, set을 이용해 고유한 화자 수 카운트
    unique_speakers = set(utterance.speaker for utterance in transcript.utterances)
    total_speakers = len(unique_speakers)

    print("-" * 50)
    print(f"✅ 총 감지된 화자 수: {total_speakers}명**")
    print("-" * 50)

    print("\n--- 화자 분리 결과 (한국어) ---")

    # Utterances(발화 목록)를 순회하며 화자별 텍스트 출력
    for utterance in transcript.utterances:
        print(f"화자 {utterance.speaker}: {utterance.text}")
    print("\n" + "="*50)

# --- 실행 부분 ---

# ⚠️ 5. 실제 파일 경로 설정 (WAV 또는 M4A 파일) ⚠️
audio_file = "/content/drive/MyDrive/final_project/회의록/1105_오전회의.m4a"

try:
    if not os.path.exists(audio_file):
        # 파일이 존재하지 않으면 명시적으로 오류 발생
        raise FileNotFoundError(f"파일을 찾을 수 없습니다: {audio_file}. Google Drive 마운트 및 경로를 확인해주세요.")

    transcribe_with_speaker_diarization_korean(audio_file)
    print("🎉 화자 분리 및 전사 작업이 성공적으로 완료되었습니다.")

# AssemblyAI 인증 오류 처리 (SDK 예외 처리 구문 수정됨)
except aai.AuthenticationError:
    print("\n🚨 인증 오류: API 키가 올바르지 않거나 만료되었습니다. 키를 확인해주세요.")

# 파일 경로 오류 처리 (파일을 찾을 수 없음)
except FileNotFoundError as e:
    print(f"\n🚨 파일 경로 오류: {e}")

# 기타 모든 예외 처리
except Exception as e:
    print(f"\n❌ 예상치 못한 오류 발생: {e}")
```
